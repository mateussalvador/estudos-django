- Para executar o `manage.py`, temos de estar dentro da pasta em que o arquivo `manage.py` está. Porém, outros comandos relacionados, por exemplo, ao ambiente virtual, devem ser executados na raiz do projeto.

# Estrutura de diretórios
- Os dois principais arquivos do Django são o `manage.py`, que nunca (normalmente, né?) é alterado. Ele carrega os utilitários de linha de comando do django. No geral, possui os mesmos comandos que `django-admin`, mas tem acesso às configurações do projeto.
- O segundo é o `settings.py`, que carrega as configurações do nosso projeto.

## settings.py
- `BASE_DIR` é uma constante que armazena o caminho para a nossa pasta raiz. 

## `urls.py`
- Um terceiro arquivo do projeto do django é o `urls.py`
- Indica as rotas que o nosso sistema acessa/responde. 
- A única url que já vem definida desde o início é o `admin/`. Ao tentar acessar via navegador, é possível que seja mostrado um erro.

## O primeiro erro do Django
- Ao criar o nosso projeto com `django-admin startproject` e em seguida rodar o servidor com `python manage.py runserver`, já temos um primeiro "erro" no terminal. 


- O erro diz "Você tem 18 migrações não aplicadas". Essas configurações não aplicadas são as dos apps admin, auth, contenttypes e sessions. Para resolver, usamos o `python manage.py migrate`, e então ele cria um banco de dados padronizado do django que já traz as configurações padrões do django. Por exemplo, configurações de usuários, etc.

# Criando um super usuário
- Usamos o comando `python manage.py createsuperuser`. Agora, com o login que foi criado podemos logar no `/admin` no servidor e já conseguimos ter acesso ao sistema de admin do django.
