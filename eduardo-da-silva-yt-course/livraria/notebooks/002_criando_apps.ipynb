# Aplicações
- Após criar um projeto, podemos criar as aplicações que estarão dentro do projeto. 
- Para criar um app, usamos o comando `python manage.py startapp nome_app`. Vamos criar a aplicação `core`.

## Listando o app
- Após criar um aplicativo, devemos passá-lo na lista `INSTALLED_APPS` no `settings.py` do Django.
