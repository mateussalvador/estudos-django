# Primeiros passos
- Mais uma vez, é importante garantir que estamos com Python 3 instalado.
- Após essa confirmação, temos de acessar a pasta do projeto e criar o ambiente virtual. 
- Executar os seguintes comandos, para instalar e atualizar:
    1. `python3 -m venv venv` # Aqui, criamos o ambiente virtual
    2. `python3 -m pip install pip setuptools wheel --upgrade`
    3. ativar o ambiente virtual com `source venv/bin/activate` ou . `venv/bin/activate`(ambos para linux)
    4. `pip install django`
    5. Conferindo a instalação e versão do django com `django-admin --version`

# Adicionando nosso projeto ao GitHub
1. Acessar o GitHub, criar o repositório
2. Abrir o terminal na pasta do projeto e usar os comandos:
    2.1. `git config --global user.name "nome"`
    2.2. `git config --global user.email "email"`
    2.3. `git config --global init.defaultBranch main`
    2.4. `git init`
    2.5. `ssh-keygen` -> Enter -> Colocar a senha da chave -> Repetir -> Enter
    2.6. `cat /caminho/da/chave-ssh.pub`
    2.7. Copiar e colar no github, em settings -> ssh and gpg keys -> new ssh key -> colar
    2.8. `git remote add origin <url para interação via ssh do github>`
3. Criar um arquivo `.gitignore`
4. Para commitar, usamos `git commit -m "mensagem"
5. `git push`
