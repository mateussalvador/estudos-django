# partials
- Uma coisa muito comum é dividir o html em partes, como por exemplo, um html para o footer, outro para o header, outro para o `<head>`, etc.
- `partials` é o nome de uma pasta que armazena partes parciais da página. Deve ficar na pasta `templates` do nosso app

## pages
- É o nome de uma pasta que armazena as pastas 'completas' (sendo inteiras ou juntas por meio de `include`)

## incluindo uma parte da página
- Para incluir o template, usamos `{% include 'recipes/partials/head.html' %}`

### Nota
- O django já encontra a pasta `templates` dentro dos apps por conta própria. Sendo assim, o caminho deve ser passado a partir dela.
