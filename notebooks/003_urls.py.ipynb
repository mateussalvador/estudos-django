# `manage.py`
- Faz os mesmos comandos do django-admin, mas o admin não sabe o caminho para o arquivo de configurações do projeto, enquanto o manage.py já vem com este caminho definido. Por isso, o comum é smepre usar `manage.py`

---

# Arquivo `urls.py`
- É o arquivo onde configuramos as rotas/caminhos da aplicação, como a página home, sobre, etc.

## `urlpatterns = []`
- Lista onde passamos os caminhos dessas rotas

### `path(rota, view)`
- A path é a função importada de `django.urls` usada para passar as rotas dentro da lista `urlpatterns`
- A função recebe dois argumentos, o primeiro sendo uma string com o caminho (**essa str NUNCA deve iniciar com /, apenas terminar**). O segundo é a `view`

### `view()`
- A `view` é, basicamente, uma função que normalmente estará dentro dos apps e que recebe como argumento `request` e retorna uma função `HttpResponse`, que deve ser importada de `django.http`
