# Arquivo `urls.py`
- É a porta de entrada da nossa aplicação. Basicamente, configuramos as rotas, como a página home, sobre, etc.

## `urlpatterns = []`
- Esta é a lista onde passamos os caminhos dessas rotas. 
- Esses caminhos são passados dentro da função `path()`

### `path(rota, view)`
- A rota é justamente o nome da página que vêm após o nosso domínio, separado por uma barra
- A view é uma função com `*args` e `**kwargs`
- **Essa função normalmente está dentro de um app!**

### Nota
- Esses caminhos podem terminar com barra (`/`), mas nunca podem iniciar com uma barra.

### Exemplo
```python
urlpatterns = [
    path("admin/", admin.site.urls),
    path("sobre/", my_view)
]
```

### view
- A função deve receber `request` como argumento. Ao acessar aquela url, o cliente (navegador) fará a requisição para o servidor nesta função função, que deverá retornar o HTTP Response.
- Podemos simplesmente resolver isso usando, por exemplo, um código do tipo:

```python
from django.contrib import admin #Já vem no urls.py
from django.urls import path #Já vem no urls.py
from django.http import HttpResponse

def my_view(request):
    return HttpResponse("Uma string qualquer para ser retornada ao navegador")

urlpatterns = [
    path("admin/", admin.site.urls), #já vem no urls.py
    path("sobre/", my_view),
]
```
