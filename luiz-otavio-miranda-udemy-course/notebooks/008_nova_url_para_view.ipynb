# URL dinâmica
- A ideia é criar uma URL dinâmica para cada receita

## Como criar uma URL dinâmica?
- Como já sabemos, para criar uma url precisamos criar uma view. 
- Para especificar que aquela é uma url dinâmica, basta colocarmos, após o caminho, um `<nome_parametro>`. Esse nome é, basicamente, o parâmetro que vai como parâmetro para a view.
- Também é preciso passar esse parâmetro para a nossa view. 

### Delimitando o tipo de dado aceito como parâmetro para a nossa URL dinâmica
- Basta informar antes do parâmetro, separando ambos por dois pontos. Ex.: `<int:nome_parametro>`

### tipos de dados aceitos como parâmetro para url dinâmica 
- São chamados de conversores de caminho
- As opções são:
    - **`str`:** Qualquer string não vazia, excluindo o separador de caminho (/)
    - **`int`:** Zero ou qualquer inteiro positivo
    - **`slug`:** Qualquer string que seja composta apenas de letras, números, hífens e underscores
    - **`uuid`:** Identificador universal único no formato padrão. Ex.: /pedido/a1b2c3d4-e5f6-7890-1234-567890abcdef/ 
    - **`path`:** qualquer string não vazia, incluindo o separador de caminho (/)

---

### Nota
- Sempre lembrar de encerrar a rota com uma barra. Ex.: `recipes/<id>/`
